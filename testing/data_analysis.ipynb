```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sunny_df = pd.read_csv("data/sunny_weather.csv")
night_df = pd.read_csv("data/night_scenario.csv")
foggy_df = pd.read_csv("data/foggy_weather.csv")
rainy_df = pd.read_csv("data/rainy_weather.csv")

df = pd.concat([sunny_df, night_df, foggy_df, rainy_df], ignore_index=True)
```

```python
summary = df.groupby("Weather").agg({
    "Distance": ["min", "max", "mean", "std"],
    "AvgConfidence": ["mean", "std"],
    "NumDetections": "sum"
}).reset_index()

print(summary)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Weather", y="Distance", data=df)
plt.title("Detection Distance Distribution by Weather Condition")
plt.ylabel("Distance (m)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Weather", y="AvgConfidence", data=df)
plt.title("Detection Confidence Distribution by Weather Condition")
plt.ylabel("Average Confidence")
plt.show()
```
